```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('company_sales_data.csv')
df.head()
```

Exercise 1: Read Total profit of all months and show it
using a line plot

```python
plt.plot(df['month_number'], df['total_profit'])
```

Exercise 2: Get total profit of all months and show line
plot with the following Style properties

```python
plt.plot(df['month_number'], df['total_units'], linestyle='--', color='red', marker='o', markerfacecolor='red', linewidth=3, label='Profit')

# Add labels and title
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.title(f'Total Profit of All Months')

# Add legend
plt.legend(loc='lower right')

# Show the plot
plt.show()
```

Exercise 3: Read all product sales data and show it
using a multiline plot


```python
plt.plot(df['month_number'], df['facecream'], label = 'face cream')
plt.plot(df['month_number'], df['facewash'], label = 'face wash')
plt.plot(df['month_number'], df['toothpaste'], label = 'toothpast')
plt.plot(df['month_number'], df['bathingsoap'], label = 'bathing soap')
plt.plot(df['month_number'], df['shampoo'], label = 'shampoo')
plt.plot(df['month_number'], df['moisturizer'], label = 'moisturizer')
plt.legend()
```

```python
plt.scatter(df['month_number'], df['toothpaste'], marker='o')
plt.grid(True)
plt.xlabel('Month')
plt.ylabel('Toothpaste')
```

Exercise 5: Read face cream and facewash product
sales data and show it using the bar chart

```python
bar_width = 0.35
plt.bar(df['month_number'], df['facecream'], width=bar_width, label='Face Cream')
plt.bar([month + bar_width for month in df['month_number']], df['facewash'], width= bar_width, label='Face Wash')

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Face Cream and Face Wash Sales per Month')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

Exercise 6: Read sales data of bathing soap of all

Exercise 7: Read the total profit of each month and
show it using the histogram to see the most common
profit ranges

```python
plt.hist(df['total_profit'], color = 'skyblue', label='Total Profit', edgecolor='black')
plt.xlabel('Profit Range')
plt.ylabel('Frequency')
plt.title('Histogram of Total Profit per Month')

# Show the plot
plt.show()
```

```python
last_year_sales = df.tail(12)-df['total_profit']

# Calculate total sales for each product
total_sales_per_product = last_year_sales.sum()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(total_sales_per_product, labels=total_sales_per_product.index, autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'lightcoral', 'orange'])

# Add title
plt.title('Total Sales Data for Last Year by Product')

# Show the plot
plt.show()
```

Exercise 9: Read Bathing soap facewash of all months
and display it using the Subplot

```python
bathing_soap_sales = df['bathingsoap']
facewash_sales = df['facewash']
months = df['month_number']

# Create subplots
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# Plot bathing soap sales
axs[0].plot(months, bathing_soap_sales, color='blue', marker='o')
axs[0].set_title('Bathing Soap Sales per Month')
axs[0].set_xlabel('Month')
axs[0].set_ylabel('Sales')

# Plot facewash sales
axs[1].plot(months, facewash_sales, color='green', marker='s')
axs[1].set_title('Facewash Sales per Month')
axs[1].set_xlabel('Month')
axs[1].set_ylabel('Sales')

# Adjust layout
plt.tight_layout()

# Show the subplots
plt.show()
```

Exercise Question 10: Read all product sales data and
show it using the stack plot

```python
products = ['facewash','facecream','bathingsoap','toothpaste' ,'shampoo','moisturizer']  # Assuming the names of the products
product_sales = [df[product] for product in products]

# Create a stack plot
plt.figure(figsize=(10, 6))
plt.stackplot(range(len(df)), product_sales, labels=products)

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Product Sales per Month')

# Add legend
plt.legend(loc='upper left')

# Show the plot
plt.show()
```
